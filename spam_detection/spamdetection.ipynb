youtube spam detection project

```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv("Youtube01-Psy.csv")
df.head()
```

```python
df['label'] = df['CLASS'].map({0: 'not spam', 1: 'spam'})
df.head()
df1 = df[['CONTENT', 'CLASS', 'label']]
df1

```

now split the data for training


```python
x=np.array(df1['CONTENT'])
y=df['label']


from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(x)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y)




```

now will transfor and train the data

```python
from sklearn.naive_bayes import BernoulliNB
model=BernoulliNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python


from sklearn.metrics import classification_report

# y_test = true labels
# y_pred = predicted labels
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

Taking input form the user and predicting the result

```python
user=input("Enter the comment : \n")
us_vec=cv.transform([user])

prediction=model.predict(us_vec)[0]

print(prediction)



```
