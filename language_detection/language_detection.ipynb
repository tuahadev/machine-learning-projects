language detection using multinomalNB

```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/dataset.csv")
df.head()
```

```python
df.isna().sum()
```

```python
x=np.array(df['Text'])
y=df['language']

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(analyzer='char', ngram_range=(1, 3)) 
vectorizer.fit(x)
x=vectorizer.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

```

Implementing MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

checking accuracy of the model

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

Asking user input to predict the language


```python
user=input("Enter the language : \n")
user_t=vectorizer.transform([user])
pred=clf.predict(user_t)

print("The language you have typed in is ",pred[0])

```

```python
from sklearn.metrics import classification_report


print(classification_report(y_test, y_pred))
```
